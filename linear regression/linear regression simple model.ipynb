```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,4,5,3,4])
y=np.array([8,7,4,5,6,4])

plt.scatter(x,y,marker='x',c='r')
plt.title("sample scatter plot")
```

```python
n=len(x)
print(n)
```

```python
m=100
c=100
y_cap=np.zeros(n)
def fuc(x,m,c):
    
    for i in range(n):
        y_cap[i]=m*x[i]+c
    return y_cap

t=fuc(x,m,c)
        
plt.scatter(x,y,marker='x',c='r', label='actual values')
plt.plot(x,t,c='b',label='our predictions')
```
