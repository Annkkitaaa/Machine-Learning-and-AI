### MORE NUMERICAL ACCURATE MODEL

```python

    # make  dataset for example
centers = [[-5, 2], [-2, -2], [1, 2], [5, -2]]
X,Y = make_blobs(n_samples=2000, centers=centers, cluster_std=1.0,random_state=30)
    
```

```python
# Adjust the output layer to have 10 units
model = Sequential([
    Dense(units=25, activation='relu'),
    Dense(units=15, activation='relu'),
    Dense(units=10, activation='linear')  # Adjusted to have 10 units
])

# Compile the model with SparseCategoricalCrossentropy loss
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), optimizer='adam')




# Fit the model
model.fit(X, Y, epochs=100)

# Predict using the model
logits = model.predict(X)
f_x = tf.nn.softmax(logits)
```
